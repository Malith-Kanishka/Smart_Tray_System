```python
import os, random, shutil
from collections import defaultdict
```

```python
images_dir = "/content/drive/MyDrive/Annotated Dataset/Images"
labels_dir = "/content/drive/MyDrive/Annotated Dataset/Labels"
output_dir = "/content/drive/MyDrive/Annotated Dataset"
```

```python
random.seed(42)

# Create folders
for split in ["train", "valid", "test"]:
    os.makedirs(f"{output_dir}/{split}/images", exist_ok=True)
    os.makedirs(f"{output_dir}/{split}/labels", exist_ok=True)

# Group files by class
class_files = defaultdict(list)

for label_file in os.listdir(labels_dir):
    if not label_file.endswith(".txt"):
        continue

    with open(os.path.join(labels_dir, label_file), "r") as f:
        class_id = f.readline().split()[0]

    image_file = label_file.replace(".txt", ".jpg")
    class_files[class_id].append((image_file, label_file))

# Split each class
for class_id, files in class_files.items():
    random.shuffle(files)

    train = files[:7]
    valid = files[7:9]
    test  = files[9:]

    for split_name, split_files in [("train", train), ("valid", valid), ("test", test)]:
        for img, lbl in split_files:
            shutil.copy(os.path.join(images_dir, img), f"{output_dir}/{split_name}/images/{img}")
            shutil.copy(os.path.join(labels_dir, lbl), f"{output_dir}/{split_name}/labels/{lbl}")
```
